# Tipos de Clusterização

```python
import pandas as pd
import matplotlib.pyplot as plt
dados = pd.read_csv('teste.csv')

plt.plot(dados.danceability,dados.instrumentalness,'o')
plt.show()
```

# 1) K-Means

```python
from sklearn import cluster
from sklearn import metrics
from scipy.spatial.distance import cdist, pdist, squareform
import numpy as np
import matplotlib.pyplot as plt


x1= dados.danceability
x2 = dados.instrumentalness

X = np.array(list(zip(x1, x2))).reshape(len(x1), 2)
kmean = cluster.KMeans(n_clusters=2).fit(X)
labels_kmean = kmean.labels_
for i in range(len(x1)):
    if labels_kmean[i]==0:
        plt.scatter(x1[i], x2[i], c='red')
    else:
        plt.scatter(x1[i], x2[i], c='blue')
plt.show()
```

# 2) Affinity Propagation

```python
S = metrics.pairwise.euclidean_distances(X)
aff_pro = cluster.AffinityPropagation().fit(S)
labels_aff = aff_pro.labels_
for i in range(len(x1)):
    if labels_aff[i]==0:
        plt.scatter(x1[i], x2[i], c='red')
    if labels_aff[i]==1:
        plt.scatter(x1[i], x2[i], c='blue')
    if labels_aff[i]==2:
        plt.scatter(x1[i], x2[i], c='orange')
    if labels_aff[i]==3:
        plt.scatter(x1[i], x2[i], c='green')
    if labels_aff[i]==4:
        plt.scatter(x1[i], x2[i], c='pink')
    if labels_aff[i]==5:
        plt.scatter(x1[i], x2[i], c='aqua')
plt.show()
```

# 3) Mean Shift

```python
ms = cluster.MeanShift(bin_seeding=True)
ms.fit(X)
labels_ms = ms.labels_
for i in range(len(x1)):
    if labels_ms[i]==0:
        plt.scatter(x1[i], x2[i], c='red')
    if labels_ms[i]==1:
        plt.scatter(x1[i], x2[i], c='blue')
    if labels_ms[i]==2:
        plt.scatter(x1[i], x2[i], c='orange')
    if labels_ms[i]==3:
        plt.scatter(x1[i], x2[i], c='green')
plt.show()
```

# 4) Spectral Clustering

```python
sc = cluster.SpectralClustering(5)
sc.fit(X)
labels_sc = sc.labels_
for i in range(len(x1)):
    if labels_sc[i]==0:
        plt.scatter(x1[i], x2[i], c='red')
    if labels_sc[i]==1:
        plt.scatter(x1[i], x2[i], c='blue')
    if labels_sc[i]==2:
        plt.scatter(x1[i], x2[i], c='orange')
    if labels_sc[i]==3:
        plt.scatter(x1[i], x2[i], c='green')
    if labels_sc[i]==4:
        plt.scatter(x1[i], x2[i], c='pink')

plt.show()       
```

# 5) Agglomerative Clustering

```python
model = cluster.AgglomerativeClustering(10).fit(X)
labels_ward=model.labels_
for i in range(len(x1)):
    if labels_ward[i]==0:
        plt.scatter(x1[i], x2[i], c='red')
    if labels_ward[i]==1:
        plt.scatter(x1[i], x2[i], c='blue')
    if labels_ward[i]==2:
        plt.scatter(x1[i], x2[i], c='orange')
    if labels_ward[i]==3:
        plt.scatter(x1[i], x2[i], c='green')
    if labels_ward[i]==4:
        plt.scatter(x1[i], x2[i], c='pink')
    if labels_ward[i]==5:
        plt.scatter(x1[i], x2[i], c='aqua')
    if labels_ward[i]==6:
        plt.scatter(x1[i], x2[i], c='yellow')
    if labels_ward[i]==7:
        plt.scatter(x1[i], x2[i], c='black')
    if labels_ward[i]==8:
        plt.scatter(x1[i], x2[i], c='darkblue')
    if labels_ward[i]==9:
        plt.scatter(x1[i], x2[i], c='purple')
    if labels_ward[i]==10:
        plt.scatter(x1[i], x2[i], c='grey')
plt.show()
```

# 6) DBSCAN

```python
D = squareform(pdist(X))
S = 1 - (D / np.max(D))
db = cluster.DBSCAN().fit(S)
labels_db = db.labels_
for i in range(len(x1)):
    if labels_db[i]==0:
        plt.scatter(x1[i], x2[i], c='red')
    if labels_db[i]==1:
        plt.scatter(x1[i], x2[i], c='blue')
    if labels_db[i]==2:
        plt.scatter(x1[i], x2[i], c='orange')
    if labels_db[i]==3:
        plt.scatter(x1[i], x2[i], c='green')
    if labels_db[i]==-1:
        plt.scatter(x1[i], x2[i], c='purple')
plt.show()
```

# 7) Birch

```python
bc = cluster.Birch(threshold=0.1).fit(X)
labels_bc = bc.labels_
for i in range(len(x1)):
    if labels_bc[i]==0:
        plt.scatter(x1[i], x2[i], c='red')
    if labels_bc[i]==1:
        plt.scatter(x1[i], x2[i], c='blue')
    if labels_bc[i]==2:
        plt.scatter(x1[i], x2[i], c='orange')
plt.show()
```

# 8) Gaussian Mixture

```python
from sklearn import mixture
gm = mixture.GaussianMixture(n_components=3).fit(X)
labels_gm = gm.predict(X)

for i in range(len(x1)):
    if labels_gm[i]==0:
        plt.scatter(x1[i], x2[i], c='red')
    if labels_gm[i]==1:
        plt.scatter(x1[i], x2[i], c='blue')
    if labels_gm[i]==2:
        plt.scatter(x1[i], x2[i], c='green')
plt.show()
```
